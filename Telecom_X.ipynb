#ðŸ“Œ ExtracciÃ³n

```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

df
```

```python
import pandas as pd
df_plano = pd.json_normalize(data)
df_plano
```

#ðŸ”§ TransformaciÃ³n

```python
df_plano.info()
```

```python
# Verifique si hay valores Ãºnicos en cada columna
for col in df_plano.columns:
  print(f"valores Ãºnicos en la columna'{col}': {df_plano[col].nunique()}")
  if df_plano[col].nunique() < 50: # Imprima los valores solamente si no hay muchos
    print(df_plano[col].unique())
    print('-' * 50)  # LÃ­nea divisoria para mejor visualizaciÃ³n
```

```python
#Duplicados
print("Numero de duplicados", df_plano.duplicated().sum())
```

```python
# Nulos
print("Numero de Nulos", df_plano.isnull().sum())
```

```python
# VacÃ­os o en blanco
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
#Cambiando la columna "account.Charges.Total" a Float

df_plano['account.Charges.Total'] = pd.to_numeric(df_plano['account.Charges.Total'], errors='coerce')
print(df_plano['account.Charges.Total'].dtype)
```

```python
# Eliminando las columnas de Churn

df_plano = df_plano[df_plano['Churn'].str.strip() != '']
print("NÃºmero de filas despuÃ©s de eliminar las vacÃ­as en 'Churn':", len(df_plano))
```

```python
# VacÃ­os o en blanco
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assuming df_plano is already loaded and processed
gender_churn = df_plano.groupby(['customer.gender', 'Churn']).size().unstack(fill_value=0)
gender_churn.plot(kind='bar', stacked=False)
plt.title('Churn por GÃ©nero')
plt.xlabel('GÃ©nero')
plt.ylabel('NÃºmero de Clientes')
plt.xticks(rotation=0)
plt.show()
```

```python
# Assuming df_plano is already loaded and processed
senior_churn = df_plano.groupby(['customer.SeniorCitizen', 'Churn']).size().unstack(fill_value=0)
senior_churn.plot(kind='bar', stacked=False)
plt.title('Churn por Senior Citizen')
plt.xlabel('Senior Citizen')
plt.ylabel('NÃºmero de Clientes')
plt.xticks(rotation=0)
plt.show()
```

```python
# Agrupar por tenure y Churn
df_grouped = df_plano.groupby(['customer.tenure', 'Churn']).size().unstack(fill_value=0)

# Calcular el % de churn por tenure
df_grouped['churn_rate'] = (df_grouped.get('Yes', 0) / df_grouped.sum(axis=1)) * 100

# Mostrar el resultado
print(df_grouped[['churn_rate']])
```

```python
# Assuming df_grouped is already calculated
plt.figure(figsize=(12, 6))
plt.plot(df_grouped.index, df_grouped['churn_rate'], marker='o')
plt.title('Tasa de Churn (%) por Tenure')
plt.xlabel('Tenure')
plt.ylabel('Tasa de Churn (%)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_plano, x="customer.tenure", y="account.Charges.Monthly", hue="Churn")
plt.title("Tenure vs Cargos mensuales por Churn")
plt.xlabel("Tenure")
plt.ylabel("Cargos mensuales")
plt.show()
```

```python
contract_churn = df_plano.groupby(['account.Contract', 'Churn']).size().unstack(fill_value=0)
contract_churn.plot(kind='bar', stacked=False)
plt.title('Churn por Contrato')
plt.xlabel('Contrato')
plt.ylabel('NÃºmero de Clientes')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df_plano, x='Churn', y='account.Charges.Monthly', hue='Churn', legend=False)
plt.title('Cargos Mensuales por Churn')
plt.xlabel('Churn')
plt.ylabel('Cargos Mensuales')
plt.show()
```
